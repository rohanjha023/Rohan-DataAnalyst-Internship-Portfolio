The goal of this task is to analyze a real-world retail sales dataset, identify data quality issues, and transform the raw data into a clean and structured format for business intelligence, dashboarding, and future advanced analytics tasks.

```python
import pandas as pd
```

### ***Data Loading***

```python
url = "https://drive.google.com/uc?export=download&id=17rL61-WJcSiEtB7fYfCDxdCzWwjtgXai"
```

```python
df = pd.read_csv(url, encoding="latin1")
df.head()
```

**Data Dictionary**

```python
data_dictionary = pd.DataFrame({
    "Column_Name": df.columns,
    "Data_Type": df.dtypes.astype(str),
    "Missing_Values": df.isnull().sum().values,
    "Unique_Values": df.nunique().values
})

data_dictionary
```

```python
data_dictionary.to_csv("task1_data_dictionary.csv", index=False)
```

**Quality Assessment**

```python
missing_values = df.isnull().sum().sort_values(ascending=False)
missing_values
```

```python
missing_percent = (df.isnull().sum() / len(df)) * 100

missing_report = pd.DataFrame({
    "Missing_Count": missing_values,
    "Missing_Percent": missing_percent
}).sort_values("Missing_Percent", ascending=False)

missing_report
```

**Cleaning**

```python
duplicate_rows = df.duplicated().sum()
print("Duplicate Rows:", duplicate_rows)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
print("Duplicates removed successfully")
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_").str.replace("-", "_")
df.columns
```

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df["delivery_days"] = (df["ship_date"] - df["order_date"]).dt.days
df["profit_margin"] = (df["profit"] / df["sales"]) * 100

df["sales_bucket"] = pd.cut(
    df["sales"],
    bins=[0, 100, 500, 1000, 5000],
    labels=["Low", "Medium", "High", "Very High"]
)

df.head()
```

```python
df.to_csv("task1_cleaned_superstore.csv", index=False)
print("Cleaned dataset saved successfully")
```
